```python
# To cc bin xp loi XL1, XL2 v XL3 da trn TBM1,TBM2 v TBM3 cho tng nm lp 
# 10, 11, 12 nh sau:
#  Nh hn 5.0 xp loi: yu (k hiu l Y)
#  T 5.0 n di 6.5: trung bnh (k hiu l TB)
#  T 6.5 n di 8.0: kh (k hiu l K)
#  T 8.0 n di 9.0: gii (k hiu l G)
#  T 9.0 tr ln: xut sc (k hiu l XS)
# df.loc[condition, "newcolumn"] = values_newcolumn
```

```python
for i in ["1","2","3"]:
    df.loc[df['TBM'+i] <= 5, 'XL'+i] = 'Y' 
    df.loc[(df['TBM'+i] <= 6.5) &(df['TBM'+i] > 5), 'XL'+i] = 'TB' 
    df.loc[(df['TBM'+i] < 8) &(df['TBM'+i] >= 6.5), 'XL'+i] = 'K' 
    df.loc[(df['TBM'+i] < 9) &(df['TBM'+i] >=8), 'XL'+i] = 'G' 
    df.loc[(df['TBM'+i] >= 9), 'XL'+i] = 'XS' 
```

```python
# To bin kt qu xt tuyn (k hiu l KQXT) nhm xc nh sinh vin u (gi tr 1) v 
# rt ( gi tr 0) vo cc khi da trn im DH1, DH2 v DH3 nh sau
#  Vi khi A, A1 nu [(DH1*2 + DH2 + DH3)/4] ln hn hoc bng 5.0 th u, 
# ngc li l rt
#  Vi khi B nu [(DH1 + DH2*2 + DH3)/4] ln hn hoc bng 5.0 th u, ngc 
# li l rt
#  Vi khi khc nu [(DH1+ DH2 + DH3)/3] ln hn hoc bng 5.0 th u, ngc 
# li l rt

for i in df["KT"].unique():
    str(i)
    df.loc[(i == "A1" or i == "A") & (((df["DH1"]*2 + df["DH2"] + df["DH3"])/4) >= 5), "KQXT"] = 1
    df.loc[(i == "A1" or i == "A") & (((df["DH1"]*2 + df["DH2"] + df["DH3"])/4) < 5), "KQXT"] = 0
    
    df.loc[(i == "B") & (((df["DH1"] + df["DH2"]*2 + df["DH3"])/4) >= 5), "KQXT"] = 1
    df.loc[(i == "B") & (((df["DH1"] + df["DH2"]*2 + df["DH3"])/4) < 5), "KQXT"] = 0
    
    df.loc[(i != "A1" or i != "A" or i == "B") & (((df["DH1"] + df["DH2"] + df["DH3"])/3) >= 5), "KQXT"] = 1
    df.loc[(i != "A1" or i != "A" or i == "B") & (((df["DH1"] + df["DH2"] + df["DH3"])/3) < 5), "KQXT"] = 0
```
